```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ComCNN_CNNBiLSTM_LSTM_greedy = [0.2874, 0.3045, 0.4028]
ComCNN_CNNBiLSTM_LSTM_beam3 = [0.2930, 0.3102, 0.4081]
ComCNN_CNNBiLSTM_LSTM_beam5 = [0.2925, 0.3093, 0.4070]
```

```python
# data to plot
n_groups = 3

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.2
opacity = 0.8

rects1 = plt.bar(index, ComCNN_CNNBiLSTM_LSTM_greedy, bar_width,
alpha=opacity,
label='Greedy search')

rects2 = plt.bar(index + bar_width, ComCNN_CNNBiLSTM_LSTM_beam3, bar_width,
alpha=opacity,
label='Beam search (k=3)')

rects3 = plt.bar(index + bar_width*2, ComCNN_CNNBiLSTM_LSTM_beam5, bar_width,
alpha=opacity,
label='Beam search (k=5)')

plt.xlabel('Metrics')
plt.ylabel('Scores')
plt.xticks(index + bar_width, ('BLEU3', 'CIDEr', 'ROUGE-L'))
plt.legend()

plt.tight_layout()
fig.set_size_inches(12,6)
fig = plt.gcf()
plt.show()
```
